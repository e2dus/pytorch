```python
import torchvision.datasets as dsets
from torchvision import transforms
import torch
import torch.utils 
import torch.nn as nn

```

```python
mnist_train = dsets.MNIST(root="MNIST_data/", train=True, transform = transforms.ToTensor(),
download= True)#root는 경로, train은 train인지 아닌지 여부, transform 변환, 파이토치에서 이미지는 0에서 1사이 값을 가지게 되고 chw순으로이루어진다.
#일반적인 이미지는 0-255사이의 값을 가지고 hwc의 순서로 되어있다.
mnist_test = dsets.MNIST(root="MNIST_data/", train=False, transform = transforms.ToTensor(),
download=True)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

train_loader = torch.utils.data.DataLoader(mnist_train,batch_size = 32,shuffle = True, drop_last = True )
test_loader = torch.utils.data.DataLoader(mnist_test, batch_size = 32,shuffle = True, drop_last = True )
training_epochs=1
for epoch in range(training_epochs):
    for X,Y in train_loader:
        X = X.view(-1, 28*28).to(device)
#X는 이미지, Y는 레이블
#X는 B C H W 로 들어오지만 베치 사이즈 체널 높이 너비 VIEW를 이용하여 B H*W로 바꾼다. 
```

```python
#epoch은 트레이닝 데이터셋이 한번돌면 1에폭이 돌았다고 한다. 
#batch size 학습을 돌때 60000장을 모두 쓸 수 없으므로 자르는데 그 크기를 얼마로 할 것 인가?
#iteration 배치를 몇번 학습에 사용햇나?

linear = torch.nn.Linear(784, 10, bias=True).to(device)
training_epochs = 15
batch_size = 100
criterion = torch.nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.SGD(linear.parameters(),lr=0.1)

for epoch in range(training_epochs):
    avg_cost = 0
    total_batch = len(train_loader)
    for X,Y in train_loader:
        X = X.view(-1, 28 * 28).to(device)
        optimizer.zero_grad()
        hypothesis = linear(X)
        cost = criterion(hypothesis, Y)
        cost.backward()
        optimizer.step()
        avg_cost += cost / total_batchㅁ
        
    print("Epoch ", epoch + 1, "cost =", "{:.9f}".format(avg_cost))

```
