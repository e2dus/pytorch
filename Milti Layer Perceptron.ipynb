```python
import torch
import torch.nn as nn
import torch.functional as F
```

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = torch.FloatTensor([[0], [1], [1], [0]])
#xor문제를 다룬다.
#nn Layers 평상시와 다르게 웨이트와 바이어스를 직접 설정해준다. 
w1 = torch.Tensor(2, 2)
b1 = torch.Tensor(2)
w2 = torch.Tensor(2, 1)
b2 = torch.Tensor(1)
#nn linear을 설정해

def sigmoid(x):
    return 1.0/(1.0 +torch.exp(-x))
def sigmoid_prime(x): #미분한 함수
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
for step in range(10001):
    l1 = torch.add(torch.matmul(X, w1), b1)
    a1 = sigmoid(l1)
    l2 = torch.add(torch.matmul(a1, w2), b2)
    Y_pred = sigmoid(l2)
    learning_rate =1
    cost = -torch.mean(Y * torch.log(Y_pred) + (1-Y) * torch.log(1 - Y_pred))
    #바이너리크로스 엔트로피 로스이다.
    
    d_Y_pred = (Y_pred - Y) / (Y_pred * (1.0 - Y_pred) +1e-7)
    #바이너리 크로스 엔트로피를 미분한식 마지막에 더한 갑슨 영으로 나누어 지는 것을 방지하기 위해서 
    
    
    #레이어 2
    d_l2 = d_Y_pred * sigmoid_prime(l2)
    d_b2 = d_l2
    d_w2 = torch.matmul(torch.transpose(a1, 0, 1), d_b2)
    #두번 째 인자 값과 3번쨰 인자 값을 바꾸어 차원축을 바꾼다. 
    
    #레이어 1
    d_a1 = torch.matmul(d_b2, torch.transpose(w2, 0, 1))
    d_l1 = d_a1 * sigmoid_prime(l1)
    d_b1 = d_l1
    d_w1 = torch.matmul(torch.transpose(X, 0, 1), d_b1)
    #backward로 해결하면 한 줄로 해결된다. 
    
    w1 = w1 - learning_rate * d_w1
    b1 = b1 - learning_rate * torch.mean(d_b1,0)
    w2 = w2 - learning_rate * d_w2
    b2 = b2 - learning_rate * torch.mean(d_b2,0)
    #step()을 사용하면 한줄로 업데이트 할 수 있다. 
    if step% 100 ==0:
        print(step, cost.item())
```

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = torch.FloatTensor([[0], [1], [1], [0]])

linear1 = torch.nn.Linear(2, 2, bias=True)
linear2 = torch.nn.Linear(2, 1, bias=True)
sigmoid = torch.nn.Sigmoid()
model = torch.nn.Sequential(linear1, sigmoid, linear2, sigmoid)
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1)
for step in range(10001):
    optimizer.zero_grad()
    hypothesis = model(X)
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()
    if step%100==0:
        print(step, cost.item())
    
```
